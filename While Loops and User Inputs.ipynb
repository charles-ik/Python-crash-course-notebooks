# Flag statements 

In a while loop, a flag variable is typically **used as a condition to determine when the loop should continue or terminate.**. 

A flag variable **is a Boolean variable** that acts as a signal or indicator within the loop. It is initially set to a certain value and is modified within the loop based on specific conditions.   
When the flag variable meets a specific condition, the loop terminates.


``` python 
flag = True  # Initializing the flag variable. this can be any variable

while flag: # basically while flag == True
    # Code inside the loop
    user_input = input("Enter a number (0 to quit): ")
    
    if user_input == '0':
        flag = False  # Changing the flag value to terminate the loop
    else:
        # Perform operations with the input
        print("You entered:", user_input)
```

# break Statement
The break statement is used to exit or terminate a loop prematurely.  
- When encountered within a loop, the break statement **immediately ends the loop execution**,    
and the **program continues with the next statement outside the loop.**

Example:

```python

while True:
    user_input = input("Enter a number (0 to quit): ")
    
    if user_input == '0':
        break  # Exit the loop if the input is '0'
    
    # Perform operations with the input
    print("You entered:", user_input)
    
```
In this example, the while True creates an infinite loop. However, when the user enters '0', the break statement is executed, breaking out of the loop and terminating the program.

# continue Statement

- used to skip the remaining code within a loop iteration and move to the next iteration. 
- When encountered within a loop, the continue statement **jumps back to the loop condition without executing the remaining statements in the loop block.**

Example:

```python
Copy code
num = 0

while num < 5:
    num += 1
    
    if num == 3:
        continue  # Skip the current iteration when num is 3
        
    print(num)
```
In this example, the while loop runs until num reaches 5. However, when num is equal to 3, the continue statement is executed, skipping the print statement for that iteration. The loop continues with the next iteration, printing the numbers 1, 2, 4, and 5.

````python
# 7-8. Deli:
# Make a list called sandwich_orders and fill it with the names of various sandwiches.
# Then make an empty list called finished_sandwiches.
# Loop through the list of sandwich orders and print a message for each order, such as I made your tuna sandwich.
# As each sandwich is made, move it to the list of finished sandwiches. 
# After all the sandwiches have been made, print a message listing each sandwich that was made.

sandwich_orders = ["Chicken Club", "BLT", "Turkey and Cheese", "Veggie Delight", "Italian Sub", "Tuna Salad"]

finished_sandwiches = []


while sandwich_orders: # basically reads as "while sandwich_orders list is NOT empty"
    being_prepared = sandwich_orders.pop() # starts from the last elemeent and take it out of the orginial list
    print(f'I am preparing you a {being_prepared} sandwich')
    finished_sandwiches.append(being_prepared)

#while loop automatically terminates as the sandwich_orders list is now empty    
    
# print all the sandwiches again

print('\n')

for sandwich in finished_sandwiches:
    print(f"I have made you a {sandwich} sandwich")


````
